Token Population:Merging the tokens from all documents, create a master list of distinct tokens available across all documents. Let us call this as “token population”

```python
stop_words = set(stopwords.words('english'))
token_pop = []
for index,r in df.iterrows():
  r['Report'] = r['Report'].lower()
  r['Report'] = nltk.tokenize.word_tokenize(r['Report'])
  for i in stop_words:
    while i in r['Report']:
      r['Report'].remove(i)
  for i in r['Report']:
    if i not in token_pop:
      token_pop.append(i)
      print(token_pop)
```

```python
bag_of_words = []
for index,r in df.iterrows():
  temporary = []
  for i in token_pop:
    if i in r['Report']:
      temporary.append(1)
    else:
      temporary.append(0)
  bag_of_words.append(temporary)

df['bag_of_words'] = bag_of_words
print("Bag of words in each student report :")
df
```

```python
count_of_words = []
for index,r in df.iterrows():
  temporary = []
  x=0
  for i in token_pop:
    if i in r['Report']:
      x=x+1
    temporary.append(x)
      
    
  count_of_words.append(temporary)

df['count_of_words'] = count_of_words

df
    
```

```python
def jaccard_set(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union
```

```python
jac_sim = []
for i in range(len(bag_of_words)):
    jc=[]
    n=0
    for j in range(len(bag_of_words)):
        n=jaccard_set(bag_of_words[i],bag_of_words[j])
        jc.append(n)
    jac_sim.append(jc)
        
x = set(bag_of_words[2])
x
```
